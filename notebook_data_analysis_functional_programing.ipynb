```python
# les imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("output_3.csv", sep="|")
df
```

```python
# ParamÃ©trage de l'affichage
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_colwidth', None)

# Afficher les premiÃ¨res lignes du dataset
print(df.head())
```

```python
# Pie chart Nutri score : RÃ©partition des produits par nutriscore
plt.figure(figsize=(8,8))
plt.title('RÃ©partition des Nutriscores', size=20)
df['nutrition_grade_fr'].value_counts(normalize=True).sort_index().plot(kind='pie', autopct='%1.0f%%')
plt.axis('equal')
plt.show()
```

```python
# Bar chart Categories : RÃ©partition des produits par catÃ©gories
plt.figure(figsize=(16,10))
plt.title('RÃ©partition des catÃ©gories (%)', size=20)
plt.xlabel('CatÃ©gories', size=15, fontweight='bold')
plt.ylabel('Proportion (%)', size=15, fontweight='bold')
df['renamed'].value_counts(normalize=True).plot(kind='bar')
plt.show()
```

```python
# SÃ©lectionner uniquement les colonnes numÃ©riques
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Calculer la matrice de corrÃ©lation sur les colonnes numÃ©riques
correlation_matrix = df[numeric_cols].corr()

# CrÃ©er le heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True)
plt.show()
```
